```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import os
import cv2
import numpy as np
```

```python
# Evaluating model
model_path='models/test_64000_9_5l.h5'
model = load_model(model_path)
input_crop='full_dataset/labelled_eval_set/crop'
input_non_crop='full_dataset/labelled_eval_set/non_crop'
threshold=[0.5,0.1,0.01,0.001]

for t in threshold:
    TP=0
    FP=0
    TN=0
    FN=0
    # Calcul recall = recognition rate in crop images
    for image in os.listdir(input_crop):
        impath=input_crop+'/'+str(image)
        img = cv2.imread(impath)
        resize = tf.image.resize(img, (256,256))
        yhat = model.predict(np.expand_dims(resize/255, 0))
        if yhat > (1-t): 
            FN=FN+1
        if yhat < t:
            TP=TP+1

    # Calculation of specificity or True Negative rate
    for image in os.listdir(input_non_crop):
        impath=input_non_crop+'/'+str(image)
        img = cv2.imread(impath)
        resize = tf.image.resize(img, (256,256))
        yhat = model.predict(np.expand_dims(resize/255, 0))
        if yhat > (1-t): 
            TN=TN+1
        if yhat < t:
            FP=FP+1


    input_file='full_dataset/random_eval_set'
    l_crop_found=[]
    l_non_crop_found=[]
    for image in os.listdir(input_file):
        impath=input_file+'/'+str(image)
        img = cv2.imread(impath)
        resize = tf.image.resize(img, (256,256))
        yhat = model.predict(np.expand_dims(resize/255, 0))
        if yhat >(1-t): 
            l_non_crop_found.append(image)
        if yhat < t:
            l_crop_found.append(image)
    precision=0
    for image in l_crop_found:
        if image in os.listdir(input_crop):
            precision=precision+1
    print('********************************************************************************************')
    print('threshold = ', t, 'model = ',model_path )
    print('precision = ',precision,'/',len(l_crop_found), ' soit ',precision/(len(l_crop_found))*100, '%') 
    print('recall = ',TP,'/',TP+FN,'  soit ',(TP/(TP+FN))*100,'% de recall')        
    print('specificité = ',TN,'/',TN+FP,'  soit ',(TN/(TN+FP))*100,'% de specificité') 
    print('Accuracy = ', TP+TN,'/',(TP+TN+FP+FN), '  soit ', ((TP+TN)/(TP+TN+FP+FN))*100,"% d'accuracy")
```
