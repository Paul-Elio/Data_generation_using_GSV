```python
images= os.listdir('labelled_crop_nigeria')
for im in images:
    angle = int(im[22:-4])
    im=im[0:22]
    filt= (df['panoid'] == im) & (df['angle']==angle) 
    df.loc[filt,'type']='crop'
```

```python
images= os.listdir('full_dataset/crop')
l_non_crop=[]
for im in images:
    angle = int(im[22:-4])
    im=im[0:22]
    if len(df[(df['panoid']== im) &(df['angle']==angle)])==0:
        l_non_crop.append([im,angle])
lost_non_crop=pd.DataFrame(l_non_crop,columns=['panoid','angle'])
lost_non_crop
```

```python
lost_point=[]
for i in range (len(df[df['type']=='collected_point'])):
    im=df[df['type']=='collected_point'].iloc[i]['panoid']
    angle=df[df['type']=='collected_point'].iloc[i]['angle']
    lost_point.append([im,angle])
df_lost_point=pd.DataFrame(lost_point,columns=['panoid','angle'])
```
