```python
# Mirror image vertically


import os
import cv2

# Path of the directory containing the images
input_dir = "full_dataset/train_data/crop"
# Path of the directory to save the augmented images
output_dir = "full_dataset/train_data/crop"

# Load the file names of the images in the input directory
image_files = os.listdir(input_dir)

# Iterate over each image
for filename in image_files:
    input_path = os.path.join(input_dir, filename)
    image = cv2.imread(input_path)
    mirrored_image = cv2.flip(image, 1)
    output_path = os.path.join(output_dir, "mirror_vert_" + filename)
    cv2.imwrite(output_path, mirrored_image)
```

```python
# Mirror image horizontaly


import os
import cv2

# Path of the directory containing the images
input_dir = "full_dataset/train_data/crop"
# Path of the directory to save the augmented images
output_dir = "full_dataset/train_data/crop"

# Load the file names of the images in the input directory
image_files = os.listdir(input_dir)

# Iterate over each image
for filename in image_files:

    input_path = os.path.join(input_dir, filename)
    image = cv2.imread(input_path)

    mirrored_image = cv2.flip(image, 0)

    output_path = os.path.join(output_dir, "mirror_hori_" + filename)
    cv2.imwrite(output_path, mirrored_image)
```

```python
# Apply random brigthness


import os
import random
from PIL import Image
import cv2
import numpy as np

# Path of the directory containing the images
input_dir = "full_dataset/train_data/crop"
# Path of the directory to save the augmented images
output_dir = "full_dataset/train_data/crop"


# Load the file names of the images in the input directory
image_files = os.listdir(input_dir)

for image_file in image_files:
    image_path = os.path.join(input_dir, image_file)
    image = Image.open(image_path)
    
    brightness_factor = random.uniform(0.8, 1.2)  
    augmented_image = image.point(lambda p: p * brightness_factor)
    augmented_image_np = np.array(augmented_image)

    output_path = os.path.join(output_dir, "bright_" + image_file)
    cv2.imwrite(output_path, augmented_image_np)
```

```python
# Aplly little rotation


import os
import random
from PIL import Image
import cv2
import numpy as np

# Path of the directory containing the images
input_dir = "full_dataset/train_data/crop"
# Path of the directory to save the augmented images
output_dir = "full_dataset/train_data/crop"

# Load the file names of the images in the input directory
image_files = os.listdir(input_dir)

for image_file in image_files:
    image_path = os.path.join(input_dir, image_file)
    image = Image.open(image_path)
    

    rotation_angle = random.randint(-10, 10)  
    augmented_image = image.rotate(rotation_angle)
    augmented_image_np = np.array(augmented_image)
    

    output_path = os.path.join(output_dir, "rotated_" + image_file)

    cv2.imwrite(output_path, augmented_image_np)
```
