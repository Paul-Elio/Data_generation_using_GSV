```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import os
import cv2
import imghdr
import numpy as np
from matplotlib import pyplot as plt
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras import layers
```

```python
model = load_model('models/test_86528_11_5l.h5')
```

```python
input_file='new_test_random'
for image in os.listdir(input_file):
    impath=input_file+'/'+str(image)
    img = cv2.imread(impath)
    resize = tf.image.resize(img, (256,256))
    yhat = model.predict(np.expand_dims(resize/255, 0))
    #if yhat > 0.99: 
        #path="full_dataset/find_non_crops/"+str(image)
        #cv2.imwrite(path, img)
        #print(image, f'Predicted class is non_crop', yhat)
    if yhat < 0.0001:
        path="test_rand/"+str(image)
        cv2.imwrite(path, img)
        print(image, f'Predicted class is crop',yhat)
```
